```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.plot(days,max_t,label='MAX')
plt.plot(days,min_t,label='MIN')
plt.plot(days,avg_t,label='AVG')
plt.legend(loc='upper left',shadow=False,fontsize='small')
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')
plt.grid()#grid is used to figure points exactly
```

```python
x = [1400,600,300,410,250]
y = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
```

```python
plt.pie(x,labels=y,shadow=False,radius=1.9,autopct='%0.0f%%',explode=[0.2,0.4,0.1,0,0.3],counterclock=True,startangle=90)
plt.axis('equal')
plt.show()
```

```python
#counterclock
#startangle
#explode
#label
#radius
#autopct
#parameters for Pie Chart
```
